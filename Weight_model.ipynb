```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
from matplotlib.ticker import MultipleLocator
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/height_weight/model_data')
```

```python
X = np.array(df['Height(Inches)']).reshape(-1, 1)
y = df['Weight(Pounds)'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=66)

lr = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(f'Training Error Percentage{round(lr.score(X_train, y_train)*100,2)}%')
print(f'Testing Error Percentage{round(lr.score(X_test, y_test)*100,2)}%')
```

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_test, y_pred)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Trained Data')
plt.show()
```
